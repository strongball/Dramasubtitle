```python
from dataset.readVideo import DramaDataset
import torchvision.transforms as transforms
transform = transforms.Compose([transforms.Resize(256),
#                                 transforms.CenterCrop(224),
#                                 transforms.ToTensor(),
#                                 transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
                                ])
import os
```

```python
DataDir="/home/ball/Videos/BrokeEN"
datasets = DramaDataset(basedir=DataDir,
                        maxFrame=1,
#                         maxSeries=1,
                        transform=transform,
                        useBmp=False
                       )
```

```python
ep = datasets.epochs[3]
```

```python
for ep in datasets.epochs:
    if (len(ep.data) < 300):
        print(ep.order, len(ep))
```

```python
for ep in datasets.epochs:
    epdir = os.path.join(datasets.frameDir, ep.order)
    if not os.path.isdir(epdir):
        os.mkdir(epdir)
    print("Make: {}".format(ep.order))
    for data in ep.data:
        filename = os.path.join(epdir, str(data["start"])+".bmp")
        imgs, sucess = ep.getFrames(data["start"], data["end"])
        if sucess:
            imgs[0].save(filename)
        else:
            print("Error")
```
